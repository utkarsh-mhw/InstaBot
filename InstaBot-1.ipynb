# <u>Follow/Unfollow given handle</u>

### open the Instagram Handle of “So Delhi”Start following it. <br>Print a message if you are already following<br>After following, unfollow the instagram handle.<br> Print a message if you have already unfollowed.

```python
def follow_profile(profile_name):
    open_profile(profile_name)
    try:
        driver.find_element_by_class_name('-fzfL')   ###unfollow button first class
        print("you are already folowing the page")
    except NoSuchElementException:
        driver.find_element_by_class_name('_6VtSN').click()  #follow button classs
        print("you have followed the page")
        
```

```python
follow_profile('sodelhi')
```

```python
def unfollow_profile(profile_name):
    open_profile(profile_name)
    try:
        driver.find_element_by_class_name('-fzfL').click()
        driver.find_element_by_class_name('-Cab_').click()
        print("you have now unfollowed the page")
    except NoSuchElementException:
        print("you have already unfollowed the handle")
    
```

# <u>Like/Unlike posts</u>

### Liking the top 30 posts of the ‘dilsefoodie'.<br> Print message if you have already liked it.<br>Unliking the top 30 posts of the ‘dilsefoodie’.<br> Print message if you have already unliked it

```python
def like_posts(profile_name, num_posts):
    open_profile(profile_name)  ## open the profile
    wait = WebDriverWait(driver, 10)  ##wait for first post
    first = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_9AhH0'))) 
    first.click()  ## open  first post  
    for i in range(num_posts):
        time.sleep(5)  ##wait
        checker = driver.find_elements_by_tag_name('svg')  ##svg label will tell if we have already liked the post previously
        if (checker[9].get_attribute('aria-label') == 'Like'):  ## means we have not liked the post
            like = driver.find_elements_by_class_name('wpO6b ')
            like[1].click()
            print("you have now liked post number "+str(i+1))
        else:                                                    ## means we have lilked the post already
            print("you had already liked post number "+str(i+1))
        wait = WebDriverWait(driver, 10)                      ## wait for next arrow button       
        ans = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_65Bje'))) ## next button  
        ans.click()  ##go to next post
```

```python
like_posts('dilsefoodie' , 30)
```

```python
def unlike_posts(profile_name, num_posts):
    open_profile(profile_name)  ## open the profile
    wait = WebDriverWait(driver, 10)  ##wait for first post
    first = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_9AhH0'))) 
    first.click()  ## open  first post  
    for i in range(num_posts):
        time.sleep(5)  ##wait
        checker = driver.find_elements_by_tag_name('svg')  ##svg label will tell if we have already liked the post previously
        if (checker[9].get_attribute('aria-label') == 'Unlike'):  ## means we have not liked the post
            like = driver.find_elements_by_class_name('wpO6b ')
            like[1].click()
            print("you have now unliked post number "+str(i+1))
        else:                                                    ## means we have lilked the post already
            print("you had already unliked post number "+str(i+1))
        wait = WebDriverWait(driver, 10)                      ## wait for next arrow button       
        ans = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_65Bje'))) ## next button  
        ans.click()  ##go to next post
```

# Extract list of followers

### Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.

```python
def get_followers(profile_name, num):
    open_profile(profile_name)
    #locate followers button
    try:
        driver.execute_script('window.scrollTo(0, 0)')
        driver.find_element_by_xpath('//a[@class="-nal3 "]/span[@class="g47SY "]').click()
    except NoSuchElementException:
        driver.find_element_by_xpath('//a[@class=" _81NM2"]/span[contains(@class, "g47SY")]').click()

    #wait for first follower to load
    waiter=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//a[contains(@class, "notranslate")]')))
    while True:
        try:
            follower_list=[]
            c=0
            while True:
                elements=driver.find_elements_by_xpath('//a[contains(@class, "notranslate")]')
    #           check if 500 followers have been extracted
                if len(elements)<num:
                    waiter=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//ul[contains(@class, "jSC57")]')))
    #             scroll for next followers
                    driver.execute_script('arguments[0].scrollIntoView(0, 100);', driver.find_element_by_xpath('//ul[contains(@class, "jSC57")]'))
                    time.sleep(0.8)
                i=elements[c]
                follower_list.append(i.get_attribute('innerHTML'))
                if len(follower_list) >= num:
                    break
                c+=1
            break
        except StaleElementReferenceException:
            continue
    return follower_list
```

```python
dilsefoodie_followers = get_followers('dilsefoodie', 500)
for i in dilsefoodie_followers:
    print(i)
```

```python
foodtlkindia_followers = get_followers('foodtalkindia', 500)
for i in foodtlkindia_followers:
    print(i)
```

### Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you

<b>Approach:</b> <br>(1)We will first get the list of people that are user's followers ->my_followers <br>(2) then we will get mutual followers of foodtalkindia (this list will have list of account that follows foodtalkindia and i follow and ):mutual_followers<br> (3)Then we will find: <center>mutual_followers - my_followers  

```python
open_profile(uname)
ans =driver.find_elements_by_class_name('g47SY ')
my_total_followers = int(ans[1].get_attribute('innerHTML'))
my_followers = get_followers(uname, my_total_followers)
```

```python
def get_mutual_followers(profile_name):
    open_profile(profile_name)
    wait = WebDriverWait(driver, 10)
    followed_by = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'tc8A9')))
    followed_by.click()
    # wait = WebDriverWait(driver, 10) 
    mutual = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'ZIAjV')))
    mutual.click()
    my_total_followers = 1    
    count = (my_total_followers//12)
    while count>=0:
        waiter=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//ul[contains(@class, "jSC57")]')))
        driver.execute_script('arguments[0].scrollIntoView(0, 100);', driver.find_element_by_xpath('//ul[contains(@class, "jSC57")]'))
        count=-1
    # wait.until(EC.presence_of_element_located((By.CLASS_NAME,'jSC57')))
    mut = driver.find_element_by_class_name('jSC57')
    time.sleep(3)
    data = mut.get_attribute('innerHTML')
    ans = BeautifulSoup(data, 'html.parser')
    ans2 = ans.find_all(class_ = 'wo9IH')
    mutual_followers= []
    for el in ans2:
        if(el.find(class_ = '_8A5w5') != None):
            mutual_followers.append(el.find(class_ = 'FPmhX')['title'])
    return mutual_followers
```

```python
mutual_followers = get_mutual_followers('foodtalkindia')
```

```python
print("followers of food talk india that i follow but those who don't follow me are:")
for i in (set(mutual_followers) - set(my_followers)):
    print(i)
```

# Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -

### If You have already seen the story.<br>Or The user has no story.<br>Or View the story if not yet seen.

```python
def see_story(profile_name):
    open_profile(profile_name)
    try:
        wait = WebDriverWait(driver , 10) 
        story = wait.until(EC.presence_of_element_located((By.XPATH, "//div[contains(@class,'h5uC0')]")))
        detail = driver.find_element_by_xpath('//div[contains(@class,"h5uC0")]/canvas')
        if (detail.get_attribute('height') == str(252)):
            print("you have not yet seen the story, it's now visible is browser")
            story.click()
        elif (detail.get_attribute('height') == str(249)):
            print("you have already seen the story of this profile")
    except TimeoutException:
        print("This profile has no story")
```

```python
see_story('coding.ninjas')
```
